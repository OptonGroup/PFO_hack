```python
import pandas as pd
import numpy as np
import os
```

```python
base_path = '/Users/savinovsvatoslav/Code/FFMOEC/train_dataset_rzhd_model_train/РЖД train/Выгрузка_маркетинговые списки'
files = [f for f in os.listdir(base_path) if f.endswith('.xls')]
```

```python
general_df = pd.DataFrame()
for folder_path in files:
    file_path = os.path.join(base_path, folder_path)
    df = pd.read_excel(file_path) # pip3 install xlrd
    df['Location'] = folder_path.split('.')[0][3:]
    general_df = pd.concat([general_df, df],axis = 0)
```

```python
general_df = general_df.fillna("undef_nan")
general_df.to_csv('all_market_lists.csv',index=False) # save to csv
x = []
y = []
for colum in general_df.columns:
    x.append(colum)
    y.append(general_df[colum].unique().shape[0])
    print(general_df[colum].unique().shape)
```

```python
# make plot with count of uniq values in every colum 
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
plt.bar(x, y, color='blue')
plt.xlabel('Columns', fontsize=12)
plt.ylabel('Unique Value Count', fontsize=12)
plt.title('Unique Values Count per Column', fontsize=14)
plt.xticks(rotation=90)
plt.yscale('log')
plt.tight_layout()
plt.show()
```
